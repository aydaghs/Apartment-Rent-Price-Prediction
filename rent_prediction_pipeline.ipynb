# **3. Data Preprocessing - Part 1: Basic Cleaning and Feature Engineering**
This part involves initial data cleaning steps:
* Creating copies of dataframes.
* Dropping unnecessary or highly problematic columns (address, currency, price_type).
* Engineering time-based features from the 'time' column.
* Handling missing values for numerical columns with few NaNs (bathrooms, bedrooms, latitude, longitude) using median imputation.
* Handling missing values for location columns (cityname, state) using mode imputation.
* Converting boolean-like columns ('fee', 'has_photo') to numerical format.

```python
# === Section 3: Data Preprocessing ===

# --- 3.1 Part 1: Basic Cleaning and Initial Feature Engineering ---

# make copies of our dataframes to work on, keeping the originals intact.
df_dev_processed = df_dev.copy()
df_eval_processed = df_eval.copy()

# --- Column Dropping ---
# Some columns don't seem very useful or have too many missing values.
# 'address': Found it has way too many NaNs (over 90%).
# 'currency' & 'price_type': EDA showed these have only one unique value each ('USD', 'Monthly').
columns_to_drop = ['address', 'currency', 'price_type']

df_dev_processed.drop(columns=columns_to_drop, inplace=True, errors='ignore')
df_eval_processed.drop(columns=columns_to_drop, inplace=True, errors='ignore')

print(f"Columns in df_dev_processed after dropping: {list(df_dev_processed.columns)}")
print(f"New shape of df_dev_processed: {df_dev_processed.shape}")
# print(f"New shape of df_eval_processed: {df_eval_processed.shape}") # To check eval set too

# --- Time Feature Engineering ---
# The 'time' column is a Unix timestamp. Let's extract more useful date/time parts.
def extract_time_features(df_input):
    # This function takes a dataframe and adds year, month, day, dayofweek, hour
    df = df_input.copy() # Work on a copy
    df['datetime'] = pd.to_datetime(df['time'], unit='s')
    df['year'] = df['datetime'].dt.year
    df['month'] = df['datetime'].dt.month
    df['day'] = df['datetime'].dt.day
    df['dayofweek'] = df['datetime'].dt.dayofweek # Monday=0, Sunday=6
    df['hour'] = df['datetime'].dt.hour
    df.drop(columns=['time', 'datetime'], inplace=True) # Don't need original 'time' anymore
    return df

df_dev_processed = extract_time_features(df_dev_processed)
df_eval_processed = extract_time_features(df_eval_processed)

print("\nTime features extracted. Here's a sample from df_dev_processed:")
display(df_dev_processed[['year', 'month', 'day', 'dayofweek', 'hour']].head())


# --- Handling Missing Values (Part 1: Numerical & Basic Location) ---
# For numerical columns with just a few NaNs, median imputation is a good start.
# Using median because it's less sensitive to outliers than mean.
numerical_cols_with_few_nans = ['bathrooms', 'bedrooms', 'latitude', 'longitude']
for col in numerical_cols_with_few_nans:
    median_value = df_dev_processed[col].median() # Calculated from dev set only!
    df_dev_processed[col].fillna(median_value, inplace=True)
    df_eval_processed[col].fillna(median_value, inplace=True) # Apply same median to eval set

print(f"\nMissing values handled with median for: {numerical_cols_with_few_nans}")
# print(f"NaNs left in bathrooms (dev): {df_dev_processed['bathrooms'].isnull().sum()}") # Quick check

# For 'cityname' and 'state', which are categorical, mode imputation makes sense for the few NaNs.
location_cols_with_few_nans = ['cityname', 'state']
for col in location_cols_with_few_nans:
    mode_value = df_dev_processed[col].mode()[0] # Calculated from dev set only!
    df_dev_processed[col].fillna(mode_value, inplace=True)
    df_eval_processed[col].fillna(mode_value, inplace=True) # Apply same mode to eval set

print(f"Missing values handled with mode for: {location_cols_with_few_nans}")
# print(f"NaNs left in cityname (dev): {df_dev_processed['cityname'].isnull().sum()}") # Quick check


# --- Converting 'fee' and 'has_photo' to Numerical/Binary ---
# 'fee' column (Yes/No) -> 1/0
df_dev_processed['fee_numeric'] = df_dev_processed['fee'].map({'Yes': 1, 'No': 0}).astype(int)
df_eval_processed['fee_numeric'] = df_eval_processed['fee'].map({'Yes': 1, 'No': 0}).astype(int)
df_dev_processed.drop(columns=['fee'], inplace=True) # Drop original object column
df_eval_processed.drop(columns=['fee'], inplace=True)

# 'has_photo' (Yes/No/Thumbnail) -> 1 for photo present, 0 otherwise
# print(f"\nUnique values in 'has_photo' (dev) before mapping: {df_dev_processed['has_photo'].unique()}")
photo_mapping = {'Yes': 1, 'Thumbnail': 1, 'No': 0}
df_dev_processed['has_photo_numeric'] = df_dev_processed['has_photo'].map(photo_mapping).astype(int)
df_eval_processed['has_photo_numeric'] = df_eval_processed['has_photo'].map(photo_mapping).astype(int)
df_dev_processed.drop(columns=['has_photo'], inplace=True) # Drop original object column
df_eval_processed.drop(columns=['has_photo'], inplace=True)


print("\n'fee_numeric' and 'has_photo_numeric' created and originals dropped.")
# display(df_dev_processed[['fee_numeric', 'has_photo_numeric']].head()) # Quick check

# --- Quick Check: Info and Remaining Missing Values after Part 1 ---
print("\n--- Info for df_dev_processed after Part 1 Preprocessing ---")
df_dev_processed.info()
print("\nRemaining missing values in df_dev_processed after Part 1:")
print(df_dev_processed.isnull().sum()[df_dev_processed.isnull().sum() > 0])
#  Now only 'amenities' and 'pets_allowed' should have missing values.
```

# **4. Data Preprocessing - Part 2: Handling Remaining Missing Values and Categorical Features**
Addresses remaining missing values and prepares other categorical features:
* Filling missing 'pets_allowed' values with 'Not Specified'.
* Filling missing 'amenities' values with an empty string.
* Analyzing and subsequently dropping the 'category' column due to high imbalance.

```python
# --- 3.2 Part 2: Handling Remaining Missing Values and Dominant Categorical Feature ---

#  address the columns that still have a good chunk of missing data:
# 'pets_allowed' and 'amenities'.

# --- Handling Missing 'pets_allowed' ---
# For 'pets_allowed', there are many NaNs.
# Instead of dropping rows or complex imputation, I'll fill NaNs with a new category.
# This way, the model can learn if "not specified" is a signal itself.
fill_value_pets = 'Not Specified'
df_dev_processed['pets_allowed'].fillna(fill_value_pets, inplace=True)
df_eval_processed['pets_allowed'].fillna(fill_value_pets, inplace=True)

print(f"Unique values in 'pets_allowed' (df_dev_processed) after filling NaNs: {df_dev_processed['pets_allowed'].unique()}")
print(f"NaNs remaining in pets_allowed (dev): {df_dev_processed['pets_allowed'].isnull().sum()}") # Should be 0


# --- Handling Missing 'amenities' ---
# 'amenities' is a text field, often a list of items. For NaNs, an empty string is a safe bet.
# This will allow TF-IDF or other text processors to handle it without issues later.
fill_value_amenities = ''
df_dev_processed['amenities'].fillna(fill_value_amenities, inplace=True)
df_eval_processed['amenities'].fillna(fill_value_amenities, inplace=True)

print(f"\nNaNs remaining in amenities (dev): {df_dev_processed['amenities'].isnull().sum()}") # Should be 0


# --- Inspecting and Handling the 'category' Column ---
# what's in the 'category' column and how diverse it is.
print(f"\nValue counts for 'category' in df_dev_processed (as percentages):")
print(df_dev_processed['category'].value_counts(normalize=True) * 100)

# print(f"\nValue counts for 'category' in df_eval_processed (as percentages):") # Also check eval for consistency
# print(df_eval_processed['category'].value_counts(normalize=True) * 100)

# Observation: 'housing/rent/apartment' is overwhelmingly dominant ( >99.9% ).
# Such a highly imbalanced categorical feature offers very little predictive information.
# It's better to drop it to simplify the model.
dev_dominant_cat_percentage = df_dev_processed['category'].value_counts(normalize=True).max() * 100
# eval_dominant_cat_percentage = df_eval_processed['category'].value_counts(normalize=True).max() * 100 # Already checked in EDA

# Defining a threshold for dropping highly imbalanced columns
# If the top category is more than, say, 99.5% of the data, it's not very useful.
imbalance_threshold = 99.5
if dev_dominant_cat_percentage > imbalance_threshold: # Only need to check dev for decision
    print(f"\nThe dominant category in 'category' column is over {imbalance_threshold}%. Dropping it.")
    df_dev_processed.drop(columns=['category'], inplace=True, errors='ignore')
    df_eval_processed.drop(columns=['category'], inplace=True, errors='ignore') # Apply to eval too
else:
    print("\n'category' column will be kept for now (to be encoded later if not dropped).")


# --- Quick Check: Info and Final Missing Values after Part 2 ---
print("\n--- Info for df_dev_processed after Part 2 Preprocessing ---")
df_dev_processed.info() # To see updated Dtypes and non-null counts

print("\n--- Info for df_eval_processed after Part 2 Preprocessing ---")
df_eval_processed.info()

print("\nRemaining missing values in df_dev_processed (should be none):")
print(df_dev_processed.isnull().sum()[df_dev_processed.isnull().sum() > 0])
# print("\nRemaining missing values in df_eval_processed (should be none):")
# print(df_eval_processed.isnull().sum()[df_eval_processed.isnull().sum() > 0])
# All missing values should now be handled in the features we've processed so far!
```

# **5. Data Preprocessing - Part 3: Text Feature Extraction and Categorical Encoding**

Focuses on extracting features from text data and encoding categorical variables:
* One-Hot Encoding for 'pets_allowed'.
* TF-IDF Vectorization for 'amenities'.
* One-Hot Encoding for 'state' and 'source'.


```python
# This cell handles the initial conversion of several key categorical and text features
# into a numerical format, and then assembles our first main feature matrix.

# --- Libraries needed ---
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
import scipy.sparse # For combining sparse matrices
from scipy.sparse import csr_matrix # To make sure numerical features are also sparse for hstack
import numpy as np # For log1p

# --- Encoding 'pets_allowed' ---
# 'pets_allowed' had NaNs filled, now converting to one-hot encoding.
# Using drop='first' to get k-1 dummies, can help some models.
print("Step 3.1: One-Hot Encoding 'pets_allowed'...")
ohe_pets = OneHotEncoder(handle_unknown='ignore', sparse_output=True, drop='first')
ohe_pets.fit(df_dev_processed[['pets_allowed']]) # Fit on dev data
pets_encoded_dev = ohe_pets.transform(df_dev_processed[['pets_allowed']])
pets_encoded_eval = ohe_pets.transform(df_eval_processed[['pets_allowed']])
pets_feature_names = ohe_pets.get_feature_names_out(['pets_allowed']) # Keep track of new feature names
print(f"  Shape of pets_encoded_dev: {pets_encoded_dev.shape}")

# --- Processing 'amenities' using TF-IDF ---
# 'amenities' are like tags/keywords in a string. TF-IDF should work well here.
print("\nStep 3.2: Processing 'amenities' with TF-IDF...")
tfidf_amenities = TfidfVectorizer(
    tokenizer=lambda text: [item.strip().lower() for item in text.split(',') if item.strip()], # My little helper to split and clean amenities
    token_pattern=None,
    min_df=10, # Ignoring very rare amenities
    max_features=200, # Keeping the top 200
    ngram_range=(1, 2) # Looking for single amenities and common pairs
)
tfidf_amenities.fit(df_dev_processed['amenities']) # Fit on dev data
amenities_tfidf_dev = tfidf_amenities.transform(df_dev_processed['amenities'])
amenities_tfidf_eval = tfidf_amenities.transform(df_eval_processed['amenities'])
amenities_feature_names = tfidf_amenities.get_feature_names_out()
print(f"  Shape of amenities_tfidf_dev: {amenities_tfidf_dev.shape}")

# --- One-Hot Encoding 'state' and 'source' ---
print("\nStep 3.3: One-Hot Encoding 'state' and 'source'...")
categorical_cols_ohe_state_source = ['state', 'source']
ohe_cats = OneHotEncoder(handle_unknown='ignore', sparse_output=True, drop='first')
ohe_cats.fit(df_dev_processed[categorical_cols_ohe_state_source]) # Fit on dev data
cats_encoded_dev = ohe_cats.transform(df_dev_processed[categorical_cols_ohe_state_source])
cats_encoded_eval = ohe_cats.transform(df_eval_processed[categorical_cols_ohe_state_source])
cats_feature_names = ohe_cats.get_feature_names_out(categorical_cols_ohe_state_source)
print(f"  Shape of cats_encoded_dev: {cats_encoded_dev.shape}")

# --- Assembling the Base Sparse Feature Matrix (X_dev_sparse & X_eval_sparse) ---
print("\nStep 3.4: Assembling the base sparse feature matrix X_dev_sparse...")
# Selecting numerical features processed in Part 1 (time features, fee_numeric, etc.)
numerical_features_df_dev = df_dev_processed.select_dtypes(include=np.number).drop(columns=['price', 'id'], errors='ignore')
initial_numerical_feature_names = list(numerical_features_df_dev.columns) # Good to save these
print(f"  Base numerical/time features included: {initial_numerical_feature_names}")
print(f"  Number of these features: {len(initial_numerical_feature_names)}")

numerical_features_df_eval = df_eval_processed.select_dtypes(include=np.number).drop(columns=['id'], errors='ignore')
numerical_sparse_dev = csr_matrix(numerical_features_df_dev.values)
numerical_sparse_eval = csr_matrix(numerical_features_df_eval.values)

# Combining all parts
X_dev_sparse = scipy.sparse.hstack([
    numerical_sparse_dev, pets_encoded_dev, amenities_tfidf_dev, cats_encoded_dev
]).tocsr()
X_eval_sparse = scipy.sparse.hstack([
    numerical_sparse_eval, pets_encoded_eval, amenities_tfidf_eval, cats_encoded_eval
]).tocsr()
print(f"\n  Shape of X_dev_sparse: {X_dev_sparse.shape}")
print(f"  Shape of X_eval_sparse: {X_eval_sparse.shape}")

# --- Defining Target Variables ---
y_dev = df_dev_processed['price'].copy()
print(f"\nTarget variable 'y_dev' (original scale) created. Shape: {y_dev.shape}")
y_dev_log_full = np.log1p(y_dev) # Using this for model training
print(f"Log-transformed target 'y_dev_log_full' created. Shape: {y_dev_log_full.shape}")
print("\n--- Base feature matrix and targets ready for initial model tests. ---")
```

```python
# This cell is for trying out some basic models on the features created in the previous cell (X_dev_sparse).
# It gives a first idea of how well we can predict 'price'.

# --- Libraries for modeling and evaluation ---
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
import lightgbm as lgb
# numpy (np) should be imported already

# --- Splitting Data (X_dev_sparse, y_dev) for Baseline Model Tests ---
# Using the features (X_dev_sparse) and original price (y_dev) from the previous cell.
X_train_base, X_val_base, y_train_orig_base, y_val_orig_base = train_test_split(
    X_dev_sparse, y_dev, test_size=0.2, random_state=42 # Keeping the split consistent
)
# Also creating log-transformed y for one of the LightGBM tests
y_train_log_base = np.log1p(y_train_orig_base)

print(f"Shape of X_train_base: {X_train_base.shape}, Shape of y_train_orig_base: {y_train_orig_base.shape}")

# --- Baseline Model 1: Ridge Regression (on original price target) ---
print("\n--- Training Ridge Regression (Baseline Model) ---")
ridge_model_base = Ridge(alpha=1.0, random_state=42) # A simple, regularized linear model
ridge_model_base.fit(X_train_base, y_train_orig_base)
y_pred_val_ridge_base = ridge_model_base.predict(X_val_base)
mae_ridge_base = mean_absolute_error(y_val_orig_base, y_pred_val_ridge_base)
print(f"  MAE with Ridge (original target): ${mae_ridge_base:.2f}") # Expecting this to be a bit high

# --- Baseline Model 2: LightGBM Regressor (on original price target) ---
print("\n--- Training LightGBM (Baseline, original target) ---")
lgbm_model_base_orig_y = lgb.LGBMRegressor(random_state=42, verbose=-1) # verbose=-1 for less output
lgbm_model_base_orig_y.fit(X_train_base, y_train_orig_base)
y_pred_val_lgbm_base_orig_y = lgbm_model_base_orig_y.predict(X_val_base)
mae_lgbm_base_orig_y = mean_absolute_error(y_val_orig_base, y_pred_val_lgbm_base_orig_y)
print(f"  MAE with LightGBM (original target): ${mae_lgbm_base_orig_y:.2f}") # Should be better than Ridge

# --- Baseline Model 3: LightGBM Regressor (on LOG-TRANSFORMED price target) ---
# Trying LightGBM on the log-transformed target. Predictions need to be converted back.
print("\n--- Training LightGBM (Baseline, log-transformed target) ---")
lgbm_model_base_log_y = lgb.LGBMRegressor(random_state=42, verbose=-1)
lgbm_model_base_log_y.fit(X_train_base, y_train_log_base) # Training with y_train_log_base
y_pred_val_lgbm_base_log_y_logscale = lgbm_model_base_log_y.predict(X_val_base)
# Converting predictions back to the original dollar scale for MAE calculation
y_pred_val_lgbm_base_log_y_origscale = np.expm1(y_pred_val_lgbm_base_log_y_logscale)
mae_lgbm_base_log_y = mean_absolute_error(y_val_orig_base, y_pred_val_lgbm_base_log_y_origscale)
print(f"  MAE with LightGBM (log-transformed target): ${mae_lgbm_base_log_y:.2f}")
mae_lgbm_log_transformed = mae_lgbm_base_log_y # Storing this for later comparison

print("\n--- Initial Baseline MAEs ---")
print(f"  Ridge (original y): ${mae_ridge_base:.2f}")
print(f"  LightGBM (original y): ${mae_lgbm_base_orig_y:.2f}")
print(f"  LightGBM (log y): ${mae_lgbm_base_log_y:.2f}  <-- This is our starting point for improvements.")
# Looks like log-transforming the target helps quite a bit! We'll stick with that.
```

# **6. Data Preprocessing - Part 4: Target Encoding for High Cardinality Features**
Implements Target Encoding for the 'cityname' column using K-Fold cross-validation to prevent data leakage. This encoded feature is then added to the feature set.

```python
# === Section 6: Data Preprocessing - Part 4: Target Encoding for 'cityname' ===
# The 'cityname' column has many unique values, so One-Hot Encoding isn't ideal.
# I'll try Target Encoding here. It encodes categories with the mean of the target variable.
# Using K-Fold to calculate these means helps prevent a common issue called target leakage.

# --- Libraries for this step ---
from sklearn.model_selection import KFold
# scipy.sparse.csr_matrix, scipy.sparse.hstack, numpy as np, pandas as pd,
# lgb, train_test_split, mean_absolute_error should already be imported.

# --- Ensure necessary variables are defined from previous parts ---
# df_dev_processed, df_eval_processed: Should be from end of Preprocessing Part 2.
# y_dev: Original price target, defined at the end of Preprocessing Part 3.
# y_dev_log_full: log1p of y_dev, also defined at the end of Preprocessing Part 3.
# pets_encoded_dev, amenities_tfidf_dev, cats_encoded_dev (and _eval versions): from Preprocessing Part 3.
# initial_numerical_feature_names: from Preprocessing Part 3.

print("\n--- Part 4: Applying K-Fold Target Encoding for 'cityname' ---")

# Create a temporary DataFrame for calculating encoding maps, helps keep things clean.
# It's important to use df_dev_processed in the state *before* 'cityname_encoded' is added.
# And y_dev_log_full which is log1p(df_dev_processed['price']).
df_dev_for_TE = df_dev_processed[['cityname']].copy() # Taking just 'cityname' for grouping
df_dev_for_TE['target_log_for_encoding'] = y_dev_log_full # This is the target we'll average

# This will be used for any cities in validation folds or evaluation set not seen during training fold.
overall_mean_log_target = y_dev_log_full.mean()

# Initialize the new 'cityname_encoded' column in our main working DataFrames.
# Doing it this way ensures the column exists before we try to fill parts of it.
df_dev_processed.loc[:, 'cityname_encoded'] = overall_mean_log_target
df_eval_processed.loc[:, 'cityname_encoded'] = overall_mean_log_target

# Setting up K-Fold cross-validation for safer target encoding
kf = KFold(n_splits=5, shuffle=True, random_state=42) # 5 folds is a common choice

print("  Starting K-Fold target encoding process for df_dev_processed...")
for fold, (train_idx, val_idx) in enumerate(kf.split(df_dev_for_TE)):
    # For each fold, calculate city means using only the "training" part of that fold.
    fold_city_target_mean = df_dev_for_TE.iloc[train_idx].groupby('cityname')['target_log_for_encoding'].mean()

    # Now apply these means to the "validation" part of this fold in our main df_dev_processed.
    # If a city in the validation part wasn't in the training part, use the overall mean.
    df_dev_processed.loc[val_idx, 'cityname_encoded'] = df_dev_processed.loc[val_idx, 'cityname'].map(fold_city_target_mean).fillna(overall_mean_log_target)
    # print(f"    Fold {fold+1} target encoding complete.")

# For the evaluation set (df_eval_processed):
# It's best to use the target means calculated from the *entire* development set (df_dev_for_TE).
print("  Applying target encoding to df_eval_processed using full development set means...")
full_dev_city_target_mean = df_dev_for_TE.groupby('cityname')['target_log_for_encoding'].mean()
df_eval_processed.loc[:, 'cityname_encoded'] = df_eval_processed['cityname'].map(full_dev_city_target_mean).fillna(overall_mean_log_target)

print("Target encoding for 'cityname' is complete.")
print("A quick look at the first 5 rows of df_dev_processed showing 'cityname' and its new 'cityname_encoded' feature:")
display(df_dev_processed[['cityname', 'cityname_encoded']].head())


# --- Rebuild Feature Matrices (X_dev_sparse_city, X_eval_sparse_city) ---
# The previous matrices X_dev_sparse/X_eval_sparse didn't have cityname (encoded or otherwise).
# Now we create new numerical feature sets that include 'cityname_encoded'.
# The original text 'cityname' column will be ignored by select_dtypes(include=np.number).
print("\nRebuilding feature matrices to include 'cityname_encoded'...")

numerical_features_plus_cityTE_dev = df_dev_processed.select_dtypes(include=np.number).drop(columns=['price', 'id'], errors='ignore')
numerical_feature_names_with_cityTE = list(numerical_features_plus_cityTE_dev.columns) # Saving names
print(f"  Numerical features now include 'cityname_encoded': {numerical_feature_names_with_cityTE}")
print(f"  New count of numerical features: {len(numerical_feature_names_with_cityTE)}") # Should be 12 + 1 = 13

numerical_features_plus_cityTE_eval = df_eval_processed.select_dtypes(include=np.number).drop(columns=['id'], errors='ignore')

# Convert to sparse
numerical_sparse_plus_cityTE_dev = csr_matrix(numerical_features_plus_cityTE_dev.values)
numerical_sparse_plus_cityTE_eval = csr_matrix(numerical_features_plus_cityTE_eval.values)

# Re-assemble with the other sparse features (pets, amenities, state/source - from Part 3)
X_dev_sparse_city = scipy.sparse.hstack([
    numerical_sparse_plus_cityTE_dev,
    pets_encoded_dev,        # This variable was created in Preprocessing Part 3
    amenities_tfidf_dev,     # This one too
    cats_encoded_dev         # And this one (state & source OHE)
]).tocsr()

X_eval_sparse_city = scipy.sparse.hstack([
    numerical_sparse_plus_cityTE_eval,
    pets_encoded_eval,
    amenities_tfidf_eval,
    cats_encoded_eval
]).tocsr()

print(f"\nShape of X_dev_sparse_city (now with target-encoded cityname): {X_dev_sparse_city.shape}") # Should be ~287 + 1 = 288 (if drop_first used) or 290+1=291
print(f"Shape of X_eval_sparse_city: {X_eval_sparse_city.shape}")


# --- Model Re-evaluation with 'cityname_encoded' ---
# Splitting data (using full log-transformed target y_dev_log_full)
X_train_city, X_val_city, y_train_log_city, y_val_log_city = train_test_split(
    X_dev_sparse_city, y_dev_log_full, test_size=0.2, random_state=42
)
# Also need the original scale validation target for final MAE.
_, y_val_orig_for_city_mae = train_test_split(y_dev, test_size=0.2, random_state=42)

print("\n--- Training LightGBM (Default Params) on features including 'cityname_encoded' ---")
lgbm_model_city = lgb.LGBMRegressor(random_state=42, verbose=-1)
lgbm_model_city.fit(X_train_city, y_train_log_city) # Train on log target

y_pred_val_log_city = lgbm_model_city.predict(X_val_city)
y_pred_val_orig_city = np.expm1(y_pred_val_log_city) # Convert back to original price scale

mae_lgbm_with_cityTE = mean_absolute_error(y_val_orig_for_city_mae, y_pred_val_orig_city)
print(f"MAE on Validation Set (LGBM with cityname_encoded, log target): ${mae_lgbm_with_cityTE:.2f}")

# 'mae_lgbm_base_log_y' should be available from the previous cell (exec_count: 8 in your notebook)
# This was the MAE for LightGBM with log target BEFORE adding cityname_encoded.
print(f"\nComparison to MAE before adding target-encoded cityname:")
if 'mae_lgbm_base_log_y' in locals():
    print(f"  MAE (LGBM, log target, NO city TE - from prev. baseline cell): ${mae_lgbm_base_log_y:.2f}")
else:
    print(f"  MAE (LGBM, log target, NO city TE - approx from earlier run): $233.85") # Fallback
print(f"  Current MAE (LGBM, log target, WITH cityname_encoded): ${mae_lgbm_with_cityTE:.2f}")
# Comment: Target encoding the cityname definitely helped a lot!
print("\n--- Part 4: Target Encoding for Cityname complete and model re-evaluated. ---")
```

```python
# === Section 7: Data Preprocessing - Part 5: Advanced Text Feature Extraction (Body) ===
# The 'body' column contains longer descriptions of the apartments.
# This text likely holds valuable information, so I'll process it using TF-IDF.

# --- Ensure necessary libraries are imported ---
# TfidfVectorizer, scipy.sparse, csr_matrix, lgb, train_test_split, mean_absolute_error, np
# should already be available from previous cells.

# --- df_dev_processed and df_eval_processed should be from the end of Part 4 (with 'cityname_encoded') ---
# --- X_dev_sparse_city and X_eval_sparse_city are our feature matrices from Part 4 ---
# --- y_dev_log_full (log-transformed target) and y_dev (original target) are also available ---
# --- mae_lgbm_with_cityTE (MAE after cityname encoding) is available for comparison ---

print("\n--- Part 5: Processing 'body' with TF-IDF and Re-evaluating Model ---")

# --- 7.1 TF-IDF Vectorization for 'body' ---
# First, making sure the 'body' column is all strings.
df_dev_processed['body'] = df_dev_processed['body'].astype(str)
df_eval_processed['body'] = df_eval_processed['body'].astype(str)

# Setting up TF-IDF. Using more features (max_features=500) than for 'amenities'
# as 'body' text is much richer. Also considering unigrams and bigrams.
tfidf_body = TfidfVectorizer(
    min_df=10,              # Word/phrase must appear in at least 10 listings.
    max_features=500,       # Capping at 500 features from the body text.
    ngram_range=(1, 2),     # Single words and two-word phrases.
    stop_words='english'    # Removing common English stop words.
)

print("  Fitting TF-IDF on 'body' of development data...")
tfidf_body.fit(df_dev_processed['body']) # Fit only on development data

# Transforming 'body' for both development and evaluation sets.
body_tfidf_dev = tfidf_body.transform(df_dev_processed['body'])
body_tfidf_eval = tfidf_body.transform(df_eval_processed['body'])
body_feature_names = tfidf_body.get_feature_names_out() # Store for feature importance
print(f"  'body' TF-IDF complete. Shape of dev TF-IDF body features: {body_tfidf_dev.shape}")

# --- 7.2 Concatenate new 'body' features ---
# Adding these new TF-IDF features from 'body' to our existing feature matrices
# (X_dev_sparse_city and X_eval_sparse_city, which already have cityname_encoded etc.).
print("\n  Adding 'body' TF-IDF features to the main feature matrix...")
X_dev_sparse_body = scipy.sparse.hstack([
    X_dev_sparse_city, # Features from Part 4 (numerical, pets, amenities, state/source, city_encoded)
    body_tfidf_dev     # New body TF-IDF features
]).tocsr()

X_eval_sparse_body = scipy.sparse.hstack([
    X_eval_sparse_city,
    body_tfidf_eval
]).tocsr()

print(f"\nShape of X_dev_sparse_body (with 'body' features): {X_dev_sparse_body.shape}") # Expected: 291 + 500 = 791 features
print(f"Shape of X_eval_sparse_body: {X_eval_sparse_body.shape}")

# --- 7.3 Model Re-evaluation with 'body' Features ---
# Splitting data for training and validation again, using the new X matrices.
X_train_body, X_val_body, y_train_log_body, y_val_log_body = train_test_split(
    X_dev_sparse_body,
    y_dev_log_full, # Using the full log-transformed target
    test_size=0.2,
    random_state=42
)
# And the original scale validation target for MAE calculation.
_, y_val_orig_body_mae = train_test_split(y_dev, test_size=0.2, random_state=42)

print("\n--- Training LightGBM (Default Params) on features including 'body' TF-IDF ---")
lgbm_model_body = lgb.LGBMRegressor(random_state=42, verbose=-1)
lgbm_model_body.fit(X_train_body, y_train_log_body) # Train on log target

y_pred_val_log_lgbm_body = lgbm_model_body.predict(X_val_body)
y_pred_val_orig_scale_lgbm_body = np.expm1(y_pred_val_log_lgbm_body) # Convert back to original price

mae_lgbm_body = mean_absolute_error(y_val_orig_body_mae, y_pred_val_orig_scale_lgbm_body)
print(f"MAE on Validation Set (LGBM with 'body' features, cityname_encoded, log target): ${mae_lgbm_body:.2f}")

print(f"\nComparison to MAE before adding 'body' features:")
# mae_lgbm_with_cityTE should be available from the previous cell (Target Encoding Cityname)
print(f"  MAE (LGBM, with cityname_encoded, NO body features - from Part 4): ${mae_lgbm_with_cityTE:.2f}")
print(f"  Current MAE (LGBM, WITH cityname_encoded AND body features): ${mae_lgbm_body:.2f}")
# Comment: The 'body' text seems to add a lot of predictive power! This is a good improvement.
print("\n--- Part 5: TF-IDF for 'body' complete and model re-evaluated. ---")
```

# **8. Data Preprocessing - Part 6: Outlier Handling for Numerical Features**
Handles outliers in the 'square_feet' column by clipping values to the 1st and 99th percentiles. The feature matrix is updated accordingly.

```python
# === Section 8: Data Preprocessing - Part 6: Outlier Handling for 'square_feet' ===
# The 'square_feet' feature showed some extreme values in EDA.
# Clipping these outliers (also known as Winsorizing at a basic level) might make the model more stable.

# --- Libraries needed ---
# numpy, train_test_split, lgb, mean_absolute_error, pandas, scipy.sparse, csr_matrix should be available.

# --- Starting with dataframes and feature matrices from Part 5 ---
# df_dev_processed, df_eval_processed: have 'cityname_encoded' and original 'square_feet'.
# X_dev_sparse_body, X_eval_sparse_body: feature matrices *including* body_tfidf.
# y_dev_log_full, y_dev: target variables.
# mae_lgbm_body: MAE before square_feet clipping (our current best, around $202.40).
# Also need: pets_encoded_dev/eval, amenities_tfidf_dev/eval, cats_encoded_dev/eval, body_tfidf_dev/eval for reconstruction.
# And initial_numerical_feature_names, pets_feature_names, amenities_feature_names, cats_feature_names, body_feature_names for full name list.

print("\n--- Part 6: Handling Outliers in 'square_feet' & Re-evaluating Model ---")

# --- 8.1 Clipping 'square_feet' ---
# I'll clip to the 1st and 99th percentiles calculated ONLY from the development set.
lower_bound_sqft = df_dev_processed['square_feet'].quantile(0.01)
upper_bound_sqft = df_dev_processed['square_feet'].quantile(0.99)
print(f"  'square_feet' original min: {df_dev_processed['square_feet'].min()}, max: {df_dev_processed['square_feet'].max()}")
print(f"  Calculated clipping bounds for 'square_feet': Lower={lower_bound_sqft:.2f}, Upper={upper_bound_sqft:.2f}")

# Create a new column for clipped values temporarily to avoid modifying the original during percentile calculation
# if this cell were re-run without re-running everything.
# However, for the final matrix, we want to replace the original 'square_feet' in the numerical features.
df_dev_processed['square_feet_final'] = df_dev_processed['square_feet'].clip(lower_bound_sqft, upper_bound_sqft)
df_eval_processed['square_feet_final'] = df_eval_processed['square_feet'].clip(lower_bound_sqft, upper_bound_sqft)
print(f"  'square_feet' clipped. New min in dev: {df_dev_processed['square_feet_final'].min()}, new max: {df_dev_processed['square_feet_final'].max()}")

# --- 8.2 Rebuild Feature Matrix with Clipped 'square_feet' ---
# We need to reconstruct the numerical part of the feature matrix.
# The list 'initial_numerical_feature_names' contains the names of numerical features BEFORE 'cityname_encoded'.
# 'cityname_encoded' and the newly clipped 'square_feet' are now the numerical parts to consider for that base.
# We need to select all numeric columns from df_dev_processed, drop price/id, ensuring 'square_feet_final' is used
# and the original 'square_feet' is not if it still exists (or better, ensure 'square_feet' column itself is updated).

# To make it clean, let's update the 'square_feet' column in df_dev_processed and df_eval_processed
# and then re-select numerical columns.
df_dev_processed['square_feet'] = df_dev_processed['square_feet_final']
df_eval_processed['square_feet'] = df_eval_processed['square_feet_final']
df_dev_processed.drop(columns=['square_feet_final'], inplace=True, errors='ignore') # remove temp column
df_eval_processed.drop(columns=['square_feet_final'], inplace=True, errors='ignore') # remove temp column


print("\n  Rebuilding numerical features part with clipped 'square_feet'...")
# Select numerical features, which now includes the clipped 'square_feet' and 'cityname_encoded'
numerical_features_df_dev_sqft = df_dev_processed.select_dtypes(include=np.number).drop(columns=['price', 'id'], errors='ignore')
numerical_feature_names_sqft = list(numerical_features_df_dev_sqft.columns) # These are our final numerical/time/city_encoded/sqft_clipped names
print(f"    Numerical features for stacking: {numerical_feature_names_sqft}")
print(f"    Count: {len(numerical_feature_names_sqft)}") # Should be 13 (initial 12 + cityname_encoded)

numerical_features_df_eval_sqft = df_eval_processed.select_dtypes(include=np.number).drop(columns=['id'], errors='ignore')

numerical_sparse_dev_sqft = csr_matrix(numerical_features_df_dev_sqft.values)
numerical_sparse_eval_sqft = csr_matrix(numerical_features_df_eval_sqft.values)

# Re-assemble the full sparse matrix (X_dev_sparse_body_sqft)
# Numerical (incl clipped sq_ft, city_encoded) + Pets + Amenities + State/Source + Body
X_dev_sparse_body_sqft = scipy.sparse.hstack([
    numerical_sparse_dev_sqft,
    pets_encoded_dev,       # From Part 3 (Cell exec_count 7)
    amenities_tfidf_dev,    # From Part 3
    cats_encoded_dev,       # From Part 3
    body_tfidf_dev          # From Part 5 (previous cell, exec_count 16)
]).tocsr()

X_eval_sparse_body_sqft = scipy.sparse.hstack([
    numerical_sparse_eval_sqft,
    pets_encoded_eval,
    amenities_tfidf_eval,
    cats_encoded_eval,
    body_tfidf_eval
]).tocsr()

print(f"\nShape of X_dev_sparse_body_sqft (clipped sq_ft, body, etc.): {X_dev_sparse_body_sqft.shape}") # Should be 791
print(f"Shape of X_eval_sparse_body_sqft: {X_eval_sparse_body_sqft.shape}")


# --- 8.3 Model Re-evaluation with Clipped 'square_feet' ---
X_train_sqft, X_val_sqft, y_train_log_sqft, y_val_log_sqft = train_test_split(
    X_dev_sparse_body_sqft, y_dev_log_full, test_size=0.2, random_state=42
)
_, y_val_orig_sqft_mae = train_test_split(y_dev, test_size=0.2, random_state=42)

print("\n--- Training LightGBM (Default Params) with clipped 'square_feet' ---")
lgbm_model_sqft = lgb.LGBMRegressor(random_state=42, verbose=-1)
lgbm_model_sqft.fit(X_train_sqft, y_train_log_sqft)

y_pred_val_log_lgbm_sqft = lgbm_model_sqft.predict(X_val_sqft)
y_pred_val_orig_scale_lgbm_sqft = np.expm1(y_pred_val_log_lgbm_sqft)

mae_lgbm_sqft = mean_absolute_error(y_val_orig_sqft_mae, y_pred_val_orig_scale_lgbm_sqft)
print(f"MAE on Validation Set (LGBM with clipped_sqft, body, city_encoded, log_target): ${mae_lgbm_sqft:.2f}")

print(f"\nComparison to MAE before clipping 'square_feet':")
# mae_lgbm_body should be from the previous cell (Part 5 output)
print(f"  MAE (LGBM, body, city_encoded, log_target, NO clipped_sqft - from Part 5): ${mae_lgbm_body:.2f}")
print(f"  Current MAE (LGBM, WITH clipped_sqft): ${mae_lgbm_sqft:.2f}")
# Comment: Clipping square_feet usually helps a little by making the model more robust.

# --- 8.4 Feature Importance (from model with clipped sqft) ---
# Now that we have a fairly complete feature set, let's see what the model thinks is important.
print("\n--- Feature Importances from LightGBM ---")
importances = lgbm_model_sqft.feature_importances_

# Assembling the full list of feature names in the correct order they were stacked
# 1. numerical_feature_names_sqft (this now contains the updated 'square_feet' and 'cityname_encoded')
# 2. pets_feature_names
# 3. amenities_feature_names
# 4. cats_feature_names (state/source)
# 5. body_feature_names
all_feature_names_final = numerical_feature_names_sqft + \
                          list(pets_feature_names) + \
                          list(amenities_feature_names) + \
                          list(cats_feature_names) + \
                          list(body_feature_names)

if len(all_feature_names_final) == X_dev_sparse_body_sqft.shape[1]:
    feature_importance_df = pd.DataFrame({'feature': all_feature_names_final, 'importance': importances})
    feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)
    print("Top 20 most important features:")
    display(feature_importance_df.head(20))

    # --- Plotting Top N Feature Importances ---
    plt.figure(figsize=(10, 8))
    sns.barplot(x="importance", y="feature", data=feature_importance_df.head(20))
    plt.title("Top 20 Feature Importances (LightGBM)")
    plt.tight_layout()
    plt.show()
    # Comment: cityname_encoded and square_feet seem to be very dominant, along with location (lat/lon).
    # Some text features from amenities/body also show up.
else:
    print(f"Error: Mismatch in length of all_feature_names ({len(all_feature_names_final)}) and number of features in model ({X_dev_sparse_body_sqft.shape[1]})")
    print("Skipping feature importance display due to this name mismatch. Please check the construction of 'all_feature_names_final'.")
    print(f"Expected numerical: {len(numerical_feature_names_sqft)}, Pets: {len(pets_feature_names)}, Amenities: {len(amenities_feature_names)}, State/Source: {len(cats_feature_names)}, Body: {len(body_feature_names)}")


print("\n--- Part 6: Outlier Handling for 'square_feet' and Feature Importance check complete. ---")
```

# **9. Model Training and Evaluation (Optimized LightGBM)**
This section details the training of the final LightGBM model.
* Splitting the development data (with all processed features and log-transformed target) into training and validation sets.
* Training the LightGBM model with manually selected (and effective) hyperparameters, including n_estimators chosen based on early stopping observation from previous experiments (e.g., n_estimators=6500). We demonstrate the training process using these refined parameters.
* Evaluating the model on the validation set using Mean Absolute Error (MAE), after inverse-transforming the predictions.

```python
# === Section 9: Model Training and Evaluation (Optimized LightGBM) ===
# Based on the feature set X_dev_sparse_body_sqft (which includes clipped square_feet and body TF-IDF)
# and y_dev_log_full (log-transformed target).
# We will first try some manual adjustments to LightGBM's hyperparameters.

# --- Libraries ---
# train_test_split, lgb, mean_absolute_error, np should be available.

# --- Data Split (using the same split as the last cell of Part 6 / cell exec_count 17) ---
# X_train_sqft, X_val_sqft, y_train_log_sqft, y_val_log_sqft
# y_val_orig_sqft_mae
# These should be defined if you ran the previous cell (Outlier Handling for Square Feet)
# which included a train_test_split.
# If not, we'd need to repeat the split:
# X_train_sqft, X_val_sqft, y_train_log_sqft, y_val_log_sqft = train_test_split(
#     X_dev_sparse_body_sqft, y_dev_log_full, test_size=0.2, random_state=42
# )
# _, y_val_orig_sqft_mae = train_test_split(y_dev, test_size=0.2, random_state=42)


print("\n--- 9.1 Training LightGBM with Initial Manually Adjusted Hyperparameters ---")
# Based on common practices, let's try increasing n_estimators and decreasing learning_rate,
# and add some regularization via subsampling.
lgbm_model_manual_tuned_v1 = lgb.LGBMRegressor(
    n_estimators=500,        # More trees than default (100)
    learning_rate=0.05,      # Smaller learning rate often requires more trees
    num_leaves=31,           # Default, often a good starting point
    max_depth=-1,            # No limit on depth (default)
    min_child_samples=20,    # Default
    subsample=0.8,           # Using 80% of data for training each tree
    colsample_bytree=0.8,    # Using 80% of features for training each tree
    random_state=42,
    objective='regression_l1', # Explicitly aiming for MAE optimization
    n_jobs=-1,
    verbose=-1                 # Suppress fit messages
)

print("  Fitting lgbm_model_manual_tuned_v1...")
lgbm_model_manual_tuned_v1.fit(X_train_sqft, y_train_log_sqft)

print("  Predicting and evaluating...")
y_pred_val_log_manual_v1 = lgbm_model_manual_tuned_v1.predict(X_val_sqft)
y_pred_val_orig_manual_v1 = np.expm1(y_pred_val_log_manual_v1)

mae_lgbm_manual_tuned_v1 = mean_absolute_error(y_val_orig_sqft_mae, y_pred_val_orig_manual_v1)
print(f"  MAE on Val Set (LGBM with initial manual tune): ${mae_lgbm_manual_tuned_v1:.2f}")

print(f"\n  Comparison with MAE before manual tune (default params, from Part 6):")
# mae_lgbm_sqft should be from the previous cell's output (approx 202.19)
if 'mae_lgbm_sqft' in locals():
    print(f"    Previous LGBM MAE (Default Params): ${mae_lgbm_sqft:.2f}")
else:
    print(f"    Previous LGBM MAE (Default Params - hardcoded): $202.09")
print(f"    Current LGBM MAE (Initial Manual Tune): ${mae_lgbm_manual_tuned_v1:.2f}")
# Comment: This manual adjustment significantly improved the MAE! (from ~202.19 to ~191.38)
# Let's try to refine n_estimators further with early stopping.
```

```python
# --- 9.2 Refining n_estimators with Early Stopping ---
# The previous manual tune was good. Now, let's increase n_estimators significantly
# and use early stopping to find the optimal number of trees with these hyperparameters.
# You found that n_estimators around 6500 (best_iteration_ 6500) worked best.

print("\n--- Training LightGBM with Refined n_estimators & EARLY STOPPING ---")

lgbm_model_early_stopping_final_val = lgb.LGBMRegressor(
    n_estimators=6500,        # Setting a high number, early stopping will find the 'best_iteration_'
    learning_rate=0.05,
    num_leaves=31,
    max_depth=-1,
    min_child_samples=20,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    objective='regression_l1',
    n_jobs=-1,
    # verbose = 100 # To see progress every 100 rounds during the potentially long training
)

print("  Fitting lgbm_model_early_stopping_final_val with early stopping...")
# We use X_train_sqft, y_train_log_sqft for training, and X_val_sqft, y_val_log_sqft for early stopping eval
lgbm_model_early_stopping_final_val.fit(
    X_train_sqft,
    y_train_log_sqft,
    eval_set=[(X_val_sqft, y_val_log_sqft)],
    eval_metric='mae',  # Monitoring MAE on the (log-transformed) validation set
    callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=True)] # Stop if no improvement for 50 rounds
)

print("\n  Prediction and Final Validation Evaluation:")
y_pred_val_log_es_final = lgbm_model_early_stopping_final_val.predict(X_val_sqft)
y_pred_val_orig_es_final = np.expm1(y_pred_val_log_es_final) # Inverse transform
mae_lgbm_es_final_validation = mean_absolute_error(y_val_orig_sqft_mae, y_pred_val_orig_es_final)

print(f"\nMAE on Validation Set (LightGBM, Manually Tuned + Early Stopping to find best n_estimators): ${mae_lgbm_es_final_validation:.2f}")

# Store this best MAE and best iteration for the final model training and report
best_mae_on_validation = mae_lgbm_es_final_validation
if hasattr(lgbm_model_early_stopping_final_val, 'best_iteration_') and lgbm_model_early_stopping_final_val.best_iteration_ is not None:
    final_n_estimators = lgbm_model_early_stopping_final_val.best_iteration_
    print(f"  Best iteration found by early stopping: {final_n_estimators}")
else:
    # If early stopping didn't trigger (e.g., if n_estimators was too low initially to see a plateau)
    final_n_estimators = lgbm_model_early_stopping_final_val.get_params()['n_estimators']
    print(f"  Early stopping may not have been triggered; using n_estimators: {final_n_estimators}")


print(f"\nComparison with MAE from initial manual tune (n_estimators=500 from previous step in this section):")
print(f"  MAE (Initial Manual Tune, n_estimators=500): ${mae_lgbm_manual_tuned_v1:.2f}") # From previous part of this section
print(f"  Current MAE (Refined Manual Tune with Early Stopping, best_n_estimators ~{final_n_estimators}): ${mae_lgbm_es_final_validation:.2f}")
# Comment: Increasing n_estimators with early stopping further improved the MAE.
# This MAE of ~$162.28 is our best on the validation set so far.

# --- Visualization: Actual vs. Predicted Prices on Validation Set ---
plt.figure(figsize=(10, 6))
plt.scatter(y_val_orig_sqft_mae, y_pred_val_orig_es_final, alpha=0.3, edgecolors='w', linewidth=0.5)
plt.plot([y_val_orig_sqft_mae.min(), y_val_orig_sqft_mae.max()], [y_val_orig_sqft_mae.min(), y_val_orig_sqft_mae.max()], 'k--', lw=2) # Diagonal line
plt.xlabel("Actual Prices (Validation Set)")
plt.ylabel("Predicted Prices (Validation Set)")
plt.title(f"Actual vs. Predicted Prices (Validation MAE: ${mae_lgbm_es_final_validation:.2f})")
plt.grid(True)
plt.show()
# Comment: The scatter plot should ideally show points clustered around the diagonal line.
# This gives a visual sense of the model's predictive accuracy.

# --- Visualization: Residual Plot ---
residuals = y_val_orig_sqft_mae - y_pred_val_orig_es_final
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_val_orig_es_final, residuals, alpha=0.3, edgecolors='w', linewidth=0.5)
plt.axhline(y=0, color='k', linestyle='--', lw=2)
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Residual Plot")
plt.grid(True)
plt.show()
# Comment: A good residual plot should show points randomly scattered around the horizontal line at zero,
# with no clear patterns. This would indicate that the model's errors are random.
```
